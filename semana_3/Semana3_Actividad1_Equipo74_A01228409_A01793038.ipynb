# Parte 1: Fundamentos de bases de datos

**1.	Fundamentos de bases de datos y para ciencia de datos.**

  Una base de datos se define como el conjunto estructura de datos guardados en la memoria de una computadora o en la nube al que se puede acceder de varias maneras. La colección y análisis de datos es una de las formas de personalizar la experiencia del usuario, de forma que se puede crear productos que se adapten a la mayoría. Las bases de datos se pueden clasificar en dos tipos: base de datos relacionales, los datos se organizan y almacenan en tablas que se pueden vincular entre sí mediante alguna relación, la forma de comunicarse e interactuar con las bases de datos relacionales es por medio de SQL; y base de datos no relacionales (NoSQL), las cuales conectan la información almacenada en categorías más que por relaciones.
  
**2.	Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos.**

  Los almacenes de datos son un conjunto de bases de datos estructuradas en el modelo ETL (Extract Transform Load) o un conjunto de bases de datos estructuradas y no estructurudas en el modelo (Extract Load Transform). El almacenamiento de datos es un sistema para el manejo de datos, cuyo propósito principal es habilitar y apoyar a las actividades de los negocios inteligentes, es decir, aquellos que aplican técnicas avanzadas para el análisis de su campo. Usualmente contienen una gran cantidad de información recolectada a por amplios periodos de tiempo, esto ayuda a centralizar y consolidar la información con el fin de utilizarla para la comprensión y predicción de diferentes situaciones. Para el científico informático William Inmon, hay cuatro características únicas que poseen los almacenes de datos: Orientación a un tema, contiene datos sobre un tema o área funcional en particular; integral, crean consistencia entre tipos de datos provenientes de diferentes fuentes; no volátiles, los datos almacenados son estables y no cambian; y variantes de tiempo, se analiza la variación en los datos a lo largo del tiempo.


# **Parte 2: Selección y limpieza de los Datos en Python**

```python
# Importar librería de pandas para manejo de DataFrames
import pandas as pd
import numpy as np
```

```python
# Lectura de base de datos
df = pd.read_csv('https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv', index_col=0)
df.head()
```

```python
# Verificar si hay valores perdidos
print(df.isnull().any())
print(df.isna().any())
```

Se observa que hay valores perdidos en la mayoria de las columnas (x2-x23, y). Es importante conocer que tipo de valor tienen estas columanas para poder hacer una limpieza adecuada.

*   X1: Amount of the given credit (NT dollar)
*   X2: Gender (1 = male; 2 = female). 
*   X3: Education (1 = graduate school; 2 = university; 3 = high school; 4 = others). 
*   X4: Marital status (1 = married; 2 = single; 3 = others). 
*   X5: Age (year).

**X6-x11 past monthly payment records (from April to September, 2005)**

**repayment status:** 
* **-1 = pay duly;** 
* **1 = payment delay for one month;** 
* **2 = payment delay for two months;** 
* **8 = payment delay for eight months;**
* **9 = payment delay for nine months and above.**

*   X6: the repayment status in September, 2005; 
*   x11: the repayment status in April, 2005

**X12-X17: Amount of bill statement (NT dollar).**
*   X12: amount of bill statement in September, 2005.
*   X17: amount of bill statement in April, 2005.

**X18-X23: Amount of previous payment (NT dollar).**

*   X18: amount paid in September, 2005
*   X23: mount paid in April, 2005

**Y: representa la salida del modelo donde 1 es aprobada y 0 no aprobada**





```python
# Función para imprimir reducción de valores en el DataFrame
def print_df_reduction(modified_df):
  # Cantidad de datos que contiene el DataFrame
  print('Tamaño de dataframe original: ', len(df))
  print('Tamaño de DataFrame al quitar filas con valores faltantes: ',
      len(modified_df))

  print('Filas removidas: ', len(df)-len(modified_df))
  print('Porcentaje removido: ',(len(df)-len(modified_df))/len(df)*100)
```

```python
# Revisar cuantos datos se perderian si se quitan las filas 
# con información faltante
drop_missing_data_rows = df.copy()
drop_missing_data_rows.dropna(inplace = True)
print('Existen valores NA: ', drop_missing_data_rows.isna().values.any())
print('Hay casillas sin valores: ', 
      drop_missing_data_rows.isnull().values.any())
print_df_reduction(drop_missing_data_rows)
```

Dado que la base de datos que se esta utilizando cuenta con un gran número de datos, quitar las filas que le falten datos solo representaría reducir el DataFrame en 0.14%, lo cual se podría decir que es insignificativo. Sin embargo, algunas columanas se podrían colocar valores generales, como en el caso de educación, en el cual se puede poner un 4 que representa otro; y estado civil, que podría ser un 3 que representa otro.

```python
filled_df = df.copy()
filled_df['X3'].fillna(value = 4, 
                       inplace = True)
filled_df['X4'].fillna(value = 3, 
                       inplace = True)
filled_df.head()
```

Revisamos que columnas tienen mayor cantidad de datos perdidos.

```python
filled_df.isna().sum()
```

Los valores faltantes del género representados por la columna X2 y X5, se pueden sustituir por la moda, así se tendrán los datos más comunes representados en los renglones que no contengan esta información.

```python
#Columnas X2 y X5 se calcula la moda de los géneros y la edad

#Se asignan las columnas que se desean analizar a una lista
columns = ['X2','X5']

#Se calculan la moda de todos los valores por columna
columns_mode = filled_df[columns].mode()

#Se imprime el resultado del promedio de todas las columnas
print('La moda de las columnas {}'.format(columns)+' es:\n{}'.format(columns_mode.loc[0,columns])+'\n')

#Se buscan todos los valores vacíos en las columnas seleccionadas y se reemplazan por la mediana
for column in columns:
  filled_df[column].fillna(value = columns_mode.iloc[0,0], inplace = True)

#Se imprime el número de columnas vacías después de realizar el proceso de sustitución
print('Los valores totales vacíos por columna son:\n\n{}'.format(filled_df.isna().sum())+'\n')

#Se despliega el encabezado del dataframe para verificar que los valores iniciales no contengan errores
filled_df.head()
```

No obstante, rellenando las columnas X3 y X4 se logró disminuir la cantidad de datos eliminados, por lo que se puede utilizar la mediana para las columnas que tienen relación como serían X6-X11

```python
#Columnas X6 - X11 se calcula la mediana del estado de pago

#Se asignan las columnas que se desean analizar a una lista
columns = ['X6', 'X7', 'X8', 'X9', 'X10', 'X11']

#Se calculan las medianas de todos los valores por columna
columns_median = filled_df[columns].median()

#Se calculan la mediana de los valores anteriores para obtener unicamente un escalar y se redondea a entero
columns_median = int(columns_median.median())

#Se imprime la mediana de todas las columnas
print('La mediana de las columnas {}'.format(columns)+' es: {}'.format(columns_median)+'\n')

#Se buscan todos los valores vacíos en las columnas seleccionadas y se reemplazan por la mediana
for column in columns:
  filled_df[column].fillna(value = columns_median, inplace = True)

#Se imprime el número de columnas vacías después de realizar el proceso de sustitución
print('Los valores totales vacíos por columna son:\n\n{}'.format(filled_df.isna().sum())+'\n')

#Se despliega el encabezado del dataframe para verificar que los valores iniciales no contengan errores
filled_df.head()
```

Realizamos lo mismo con las columnas X12-X17, pero en vez de utilizar la mediana se utilizará el promedio debedido a que corresponde a cantidades de dinero

```python
#Columnas X2 - X17 se calcula promedio de las deudas

#Se asignan las columnas que se desean analizar a una lista
columns = ['X12', 'X13', 'X14', 'X15', 'X16', 'X17']

#Se calculan los promedios de todos los valores por columna
columns_mean = filled_df[columns].mean()

#Se calculan el promedio de los valores anteriores para obtener unicamente un escalar y se redondea a entero
columns_mean = int(columns_mean.mean())

#Se imprime el resultado del promedio de todas las columnas
print('El promedio de las columnas {}'.format(columns)+' es: {}'.format(columns_mean)+'\n')

#Se buscan todos los valores vacíos en las columnas seleccionadas y se reemplazan por la mediana
for column in columns:
  filled_df[column].fillna(value = columns_mean, inplace = True)

#Se imprime el número de columnas vacías después de realizar el proceso de sustitución
print('Los valores totales vacíos por columna son:\n\n{}'.format(filled_df.isna().sum())+'\n')

#Se despliega el encabezado del dataframe para verificar que los valores iniciales no contengan errores
filled_df.head()
```

**Se aplica el mismo criterio de sustituir por el promedio los datos faltantes en las columnas X18-X23**

```python
#Columnas X18 - X23 se calcula promedio de los pagos

#Se asignan las columnas que se desean analizar a una lista
columns = ['X18', 'X19', 'X20', 'X21', 'X22', 'X23']

#Se calculan los promedios de todos los valores por columna
columns_mean = filled_df[columns].mean()

#Se calculan el promedio de los valores anteriores para obtener unicamente un escalar y se redondea a entero
columns_mean = int(columns_mean.mean())

#Se imprime el resultado del promedio de todas las columnas
print('El promedio de las columnas {}'.format(columns)+' es: {}'.format(columns_mean)+'\n')

#Se buscan todos los valores vacíos en las columnas seleccionadas y se reemplazan por la mediana
for column in columns:
  filled_df[column].fillna(value = columns_mean, inplace = True)

#Se imprime el número de columnas vacías después de realizar el proceso de sustitución
print('Los valores totales vacíos por columna son:\n\n{}'.format(filled_df.isna().sum())+'\n')

#Se despliega el encabezado del dataframe para verificar que los valores iniciales no contengan errores
filled_df.head()
```

Sólo restan 3 datos perdidos los cuales corresponden al resultado de la aprobación o no del crédito. Se deciden remover estas filas porque no se desea hacer ninguna conclusión sobre los resultados ya que es la variable dependiente.

```python
#Columnas Y se eliminan las filas sin el resultado de la variable dependiente

filled_df.dropna(inplace = True)

#Se imprime el número de columnas vacías después de realizar el proceso de sustitución
print('Los valores totales vacíos por columna son:\n\n{}'.format(filled_df.isna().sum())+'\n')

#Se despliega el encabezado del dataframe para verificar que los valores iniciales no contengan errores
filled_df.head()
```

Al no haber datos vacios y solo haber eliminado 3 filas de toda el dataset se puede decir que se hizo una buena limpieza de datos.

```python
print('La longitud original de la base de datos era: {}'.format(len(df))+'\n')
print('La longitud de la base de datos despues de la limpieza es: {}'.format(len(filled_df))+'\n')
print('Se removieron sólo: {}'.format(len(df)-len(filled_df))+' filas de datos\n')
print('Correspondiente al: {}'.format((len(df)-len(filled_df))/len(df)*100)+'% de los datos')
```
